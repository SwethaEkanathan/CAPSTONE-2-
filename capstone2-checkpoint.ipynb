# Study Of Physical Activities and Calories burned

1: Activity Levels and Caloric Expenditure:

**Null Hypothesis:** There is no significant relationship between VeryActiveHours, FairlyActiveHours, LightlyActiveHours, and Calories.

**Alternative Hypothesis:** Different activity levels have a significant impact on caloric expenditure.

2: Sedentary Behavior and Overall Health:

**Null Hypothesis:** SedentaryHours do not affect overall health, measured by TotalDistance_km and Calories.

**Alternative Hypothesis:** Increased sedentary hours are associated with lower TotalDistance_km and Calories.

3: Distance and Very Active Hours:

**Null Hypothesis:** There is no relationship between TotalDistance_km and VeryActiveHours.

**Alternative Hypothesis:** Individuals who cover more distance tend to have more VeryActiveHours.

4: Fairly Active Hours and Caloric Expenditure:

**Null Hypothesis:** FairlyActiveHours have no impact on Calories.

**Alternative Hypothesis:** There is a positive correlation between FairlyActiveHours and Calories.

5: Weekday vs. Weekend Activity:

**Null Hypothesis:** There is no significant difference in activity levels (VeryActiveHours, FairlyActiveHours, LightlyActiveHours) between weekdays and weekends.

**Alternative Hypothesis:** Activity patterns vary significantly between weekdays and weekends.

6: Change in Activity Levels Over Time:

**Null Hypothesis:** There is no significant change in activity levels over the observed period.

**Alternative Hypothesis:** Activity levels show a significant trend or pattern over time.

7: Impact of Very Active Hours on Total Distance:

**Null Hypothesis:** VeryActiveHours have no impact on the total distance covered.

**Alternative Hypothesis:** Individuals with more VeryActiveHours cover a greater total distance.

8: Sedentary Hours and Total Distance:

**Null Hypothesis:** SedentaryHours do not affect the total distance covered.

**Alternative Hypothesis:** Increased sedentary hours are associated with a decrease in the total distance covered.

9: Correlation between Different Activity Levels:

**Null Hypothesis:** There is no correlation between VeryActiveHours, FairlyActiveHours, and LightlyActiveHours.

**Alternative Hypothesis:** There is a significant correlation between different activity levels.

10: Gender Differences in Activity Patterns:

**Null Hypothesis:** There is no significant difference in activity patterns between genders.

**Alternative Hypothesis:** There are gender-based variations in activity levels and patterns.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_csv('data/Activities.csv')
df.head()
```

```python
# Summary statistics of the dataset
df.describe()
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Visualize the distribution of numeric variables
plt.figure(figsize=(16, 10))
sns.pairplot(df)
plt.show()
```


The pairplot shows the relationship between the numeric variables in the dataset. The diagonal plots show the distribution of each variable, and the off-diagonal plots show the relationship between each pair of variables.

The distribution of VeryActiveHours, FairlyActiveHours, and LightlyActiveHours is skewed to the right, indicating that most people have low levels of these activities. The distribution of Calories is also skewed to the right, but it is not as pronounced as the distribution of the activity levels.

The relationship between VeryActiveHours, FairlyActiveHours, and LightlyActiveHours is positive, indicating that people who do more of one activity tend to do more of the others. The relationship between Calories and the activity levels is also positive, but it is not as strong as the relationship between the activity levels themselves.

The relationship between SedentaryHours and TotalDistance_km is negative, indicating that people who spend more time being sedentary tend to cover less distance. The relationship between SedentaryHours and Calories is also negative, but it is not as strong as the relationship between SedentaryHours and TotalDistance_km.

The relationship between TotalDistance_km and VeryActiveHours is positive, indicating that people who cover more distance tend to have more VeryActiveHours. The relationship between FairlyActiveHours and Calories is also positive, but it is not as strong as the relationship between TotalDistance_km and VeryActiveHours.

The relationship between VeryActiveHours and TotalDistance is positive, indicating that people with more VeryActiveHours cover a greater total distance. The relationship between SedentaryHours and TotalDistance is negative, indicating that increased sedentary hours are associated with a decrease in the total distance covered.

There is a significant correlation between VeryActiveHours, FairlyActiveHours, and LightlyActiveHours, indicating that people who do more of one activity tend to do more of the others. There is also a significant correlation between SedentaryHours and TotalDistance_km, indicating that people who spend more time being sedentary tend to cover less distance.

There is no significant difference in activity patterns between genders, indicating that men and women have similar levels of activity.

```python
# Activity Levels and Caloric Expenditure
corr_activity_calories, p_activity_calories = stats.pearsonr(df['VeryActiveHours'] + df['FairlyActiveHours'] + df['LightlyActiveHours'], df['Calories'])
if p_activity_calories < 0.05:
    print("Reject Null Hypothesis: Different activity levels have a significant impact on caloric expenditure.")
else:
    print("Fail to Reject Null Hypothesis: No significant relationship between activity levels and caloric expenditure.")
```

```python
sns.regplot(x=df['VeryActiveHours'] + df['FairlyActiveHours'] + df['LightlyActiveHours'], y=df['Calories'])
plt.xlabel('Total Active Hours')
plt.ylabel('Calories')
plt.title('Activity Levels vs. Caloric Expenditure')
plt.show()
```

The chart shows the relationship between total active hours and calories. As total active hours increase, calories also increase. This is consistent with the hypothesis that different activity levels have a significant impact on caloric expenditure.

```python
sns.violinplot(x=df['VeryActiveHours'] + df['FairlyActiveHours'] + df['LightlyActiveHours'], y=df['Calories'])
plt.xlabel('Total Active Hours')
plt.ylabel('Calories')
plt.title('Activity Levels vs. Caloric Expenditure')
plt.show()
```

The violin plot shows the distribution of calories for different levels of total active hours. The plot shows that as total active hours increase, the distribution of calories also increases. This is consistent with the hypothesis that different activity levels have a significant impact on caloric expenditure.

```python
# Sedentary Behavior and Overall Health
corr_sedentary_health, p_sedentary_health = stats.pearsonr(df['SedentaryHours'], df['TotalDistance_km'] + df['Calories'])
if p_sedentary_health < 0.05:
    print("Reject Null Hypothesis: Increased sedentary hours are associated with lower TotalDistance_km and Calories.")
else:
    print("Fail to Reject Null Hypothesis: No significant relationship between sedentary hours and overall health.")
```

```python
sns.pairplot(df[['SedentaryHours', 'TotalDistance_km', 'Calories']])
plt.suptitle('Pair Plot: Sedentary Behavior and Overall Health')
plt.show()
```

The pairplot shows the relationship between sedentary hours, total distance covered, and calories burned. The diagonal plots show the distribution of each variable, and the off-diagonal plots show the relationship between each pair of variables.

The distribution of sedentary hours is skewed to the right, indicating that most people spend a relatively small amount of time being sedentary. The distribution of total distance covered is also skewed to the right, but it is not as pronounced as the distribution of sedentary hours. The distribution of calories burned is skewed to the left, indicating that most people burn a relatively small number of calories.

The relationship between sedentary hours and total distance covered is negative, indicating that people who spend more time being sedentary tend to cover less distance. The relationship between sedentary hours and calories burned is also negative, but it is not as strong as the relationship between sedentary hours and total distance covered.

The relationship between total distance covered and calories burned is positive, indicating that people who cover more distance tend to burn more calories.

The correlation between sedentary hours and total distance covered is significant, indicating that there is a relationship between these two variables. The correlation between sedentary hours and calories burned is also significant, indicating that there is a relationship between these two variables.

The correlation between total distance covered and calories burned is also significant, indicating that there is a relationship between these two variables.

Overall, the plots suggest that sedentary behavior is associated with lower total distance covered and calories burned.

```python
sns.boxplot(data=df[['SedentaryHours', 'TotalDistance_km', 'Calories']])
plt.title('Box Plot: Sedentary Behavior and Overall Health')
plt.show()
```


The chart suggests that most people spend a relatively small amount of time being sedentary, cover a relatively small distance, and burn a relatively small number of calories. However, there is a significant amount of variation in each of these variables.

```python
#Distance and Very Active Hours
corr_distance_very_active, p_distance_very_active = stats.pearsonr(df['TotalDistance_km'], df['VeryActiveHours'])
if p_distance_very_active < 0.05:
    print("Reject Null Hypothesis: Individuals who cover more distance tend to have more VeryActiveHours.")
else:
    print("Fail to Reject Null Hypothesis: No significant relationship between distance and VeryActiveHours.")
```

```python
sns.regplot(x=df['TotalDistance_km'], y=df['VeryActiveHours'])
plt.xlabel('Total Distance (km)')
plt.ylabel('Very Active Hours')
plt.title('Distance vs. Very Active Hours')
plt.show()
```


The chart shows the relationship between total distance covered and very active hours. As total distance covered increases, very active hours also increase. This is consistent with the hypothesis that individuals who cover more distance tend to have more VeryActiveHours.

The correlation between total distance covered and very active hours is significant, indicating that there is a relationship between these two variables. The slope of the regression line is positive, indicating that as total distance covered increases, very active hours also increase.

The chart also shows that there is some variability in the relationship between total distance covered and very active hours. Some individuals who cover a lot of distance do not have a lot of very active hours, while others who cover a moderate amount of distance have a lot of very active hours. This variability is likely due to a number of factors, such as individual fitness level, activity preferences, and access to resources.


```python
sns.jointplot(x=df['TotalDistance_km'], y=df['VeryActiveHours'], kind='reg')
plt.xlabel('Total Distance (km)')
plt.ylabel('Very Active Hours')
plt.title('Distance vs. Very Active Hours')
plt.show()
```

```python
# Fairly Active Hours and Caloric Expenditure
corr_fairly_active_calories, p_fairly_active_calories = stats.pearsonr(df['FairlyActiveHours'], df['Calories'])
if p_fairly_active_calories < 0.05:
    print("Reject Null Hypothesis: There is a positive correlation between FairlyActiveHours and Calories.")
else:
    print("Fail to Reject Null Hypothesis: FairlyActiveHours have no impact on Calories.")
```

```python
sns.regplot(x=df['FairlyActiveHours'], y=df['Calories'])
plt.xlabel('Fairly Active Hours')
plt.ylabel('Calories')
plt.title('Fairly Active Hours vs. Caloric Expenditure')
plt.show()
```

```python
# Impact of Very Active Hours on Total Distance
result_va_distance = stats.linregress(df['VeryActiveHours'], df['TotalDistance_km'])
if result_va_distance.pvalue < 0.05:
    print("Reject Null Hypothesis: VeryActiveHours have a significant impact on the total distance covered.")
else:
    print("Fail to Reject Null Hypothesis: VeryActiveHours have no impact on the total distance covered.")
```

```python
sns.regplot(x=df['VeryActiveHours'], y=df['TotalDistance_km'])
plt.xlabel('Very Active Hours')
plt.ylabel('Total Distance (km)')
plt.title('Impact of Very Active Hours on Total Distance')
plt.show()
```

```python
# Sedentary Hours and Total Distance
result_sedentary_distance = stats.linregress(df['SedentaryHours'], df['TotalDistance_km'])
if result_sedentary_distance.pvalue < 0.05:
    print("Reject Null Hypothesis: Increased sedentary hours are associated with a decrease in the total distance covered.")
else:
    print("Fail to Reject Null Hypothesis: SedentaryHours do not affect the total distance covered.")
```

```python
sns.regplot(x=df['SedentaryHours'], y=df['TotalDistance_km'])
plt.xlabel('Sedentary Hours')
plt.ylabel('Total Distance (km)')
plt.title('Sedentary Hours vs. Total Distance')
plt.show()
```

```python
plt.scatter(x=df['SedentaryHours'], y=df['TotalDistance_km'], c=df['SedentaryHours'], cmap='viridis')
plt.xlabel('Sedentary Hours')
plt.ylabel('Total Distance (km)')
plt.title('Sedentary Hours vs. Total Distance')
plt.colorbar(label='Sedentary Hours')
plt.show()
```


The chart shows the relationship between sedentary hours and total distance covered. As sedentary hours increase, total distance covered decreases. This is consistent with the hypothesis that increased sedentary hours are associated with a decrease in physical activity, which in turn leads to a decrease in total distance covered.
The chart also shows that there is a lot of variation in the relationship between sedentary hours and total distance covered. Some people who spend a lot of time being sedentary still cover a lot of distance, while others who spend a relatively small amount of time being sedentary cover very little distance. This variation is likely due to a number of factors, such as individual fitness level, activity preferences, and access to resources.
Overall, the chart suggests that there is a relationship between sedentary hours and total distance covered, but that this relationship is not linear.


```python
# Correlation between Different Activity Levels
corr_activity_levels, p_activity_levels = stats.pearsonr(df['VeryActiveHours'], df['FairlyActiveHours'])
if p_activity_levels < 0.05:
    print("Reject Null Hypothesis: There is a significant correlation between VeryActiveHours and FairlyActiveHours.")
else:
    print("Fail to Reject Null Hypothesis: No correlation between VeryActiveHours and FairlyActiveHours.")
```

```python
sns.pairplot(df[['VeryActiveHours', 'FairlyActiveHours']])
plt.suptitle('Pair Plot: Correlation between Activity Levels')
plt.show()
```

```python
correlation_matrix = df[['VeryActiveHours', 'FairlyActiveHours']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap: Correlation between Activity Levels')
plt.show()
```
